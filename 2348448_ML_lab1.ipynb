# **EDA USING PARAMETRIC METHODS**

## **Descriptive Statistics**

*   the skewness is negative, the distribution is skewed to the left
*   If the kurtosis is positive, the distribution has
heavier tails and a sharper peak than the normal distribution. This is called leptokurtic.
*   If the kurtosis is negative, the distribution has lighter tails and a flatter peak than the normal distribution. This is called platykurtic.
*   If the kurtosis is close to zero, the distribution is similar to the normal distribution and is called mesokurtic.
  





```python
data=df.Profit
mean = np.mean(data)
std_dev = np.std(data)
skewness = np.mean((data - mean) ** 3) / (std_dev ** 3)
kurtosis = np.mean((data - mean) ** 4) / (std_dev ** 4) - 3

print("Descriptive Statistics:")
print(f"Mean: {mean:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
print(f"Skewness: {skewness:.2f}")
print(f"Kurtosis: {kurtosis:.2f}")
```

## **Histogram**

## **Test Statistic(t-test)**
The test statistic indicates how well the data conforms to a normal distribution.

```python
coffee_sales = df[df['Product Type'] == 'Coffee']['Sales']
tea_sales = df[df['Product Type'] == 'Tea']['Sales']

t_statistic, p_value = stats.ttest_ind(coffee_sales, tea_sales)

print("\nT-test results for sales between coffee and tea:")
print("T-statistic:", t_statistic)
print("P-value:", p_value)
```

## **QQ Plot**

* Create a QQ plot with the empirical quantiles of data on the y-axis and the theoretical quantiles of a standard normal distribution on the x-axis.
* The line='s' argument specifies that a line representing the theoretical quantiles of the standard normal distribution should be added to the plot.
* The plot allows for visual inspection of whether the data points fall close to the diagonal line, which indicates conformity to a normal distribution.
* If the points deviate significantly from the line, it suggests departures from normality.

```python
# Probability plots - Example: Q-Q plot for coffee sales
print("\nProbability plot for coffee_sales:")
stats.probplot(coffee_sales, dist="norm", plot=plt)
plt.title("Q-Q plot for Coffee Sales")
plt.show()
```

## **Normal Distribution**

Fit a normal distribution

```python
mu, sigma = stats.norm.fit(tea_sales)
print("\nParameters of the fitted normal distribution for tea sales:")
print("Mean (mu):", mu)
print("Standard deviation (sigma):", sigma)
```

```python
plt.hist(tea_sales, bins=5, density=True, alpha=0.6, color='g', label='tea_sales')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mu, sigma)
plt.plot(x, p, 'k', linewidth=2, label='Fitted Normal Distribution')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Histogram and Fitted Normal Distribution for Tea Sales')
plt.legend()
plt.show()
```
